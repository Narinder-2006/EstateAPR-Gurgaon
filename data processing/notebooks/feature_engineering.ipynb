```python
def checker(row):
    if "0 to 1 Year Old"   in row or "Within 3 months" in row  or "Within 6 months" in row:
        return "New Property"
    if "5 to 10 Year Old" in row:
        return "Moderately Old"
    if "10+ Year Old" in row:
        return "Old Property"
    if "1 to 5 Year Old" in row:
        return "Relatively New"
   
    if "undefined" in row or "agePossession" in row:
        return "undefined"
    else:
        return "Under Construction"
    
```

```python
df3['agePos']=df3["agePossession"].apply(checker)
```

```python
# Extract all unique furnishings from the furnishDetails column
all_furnishings = []
for detail in df3['furnishDetails'].dropna():
    furnishings = detail.replace('[', '').replace(']', '').replace("'", "").split(', ')
    all_furnishings.extend(furnishings)
unique_furnishings = list(set(all_furnishings))
```

```python
unique_furnishings
```

```python
import re
cols=[re.sub(r"No|\d+",'',Fi).strip() for Fi in unique_furnishings]
```

```python
cols=list(set(cols))
```

```python
cols=[furnishing for furnishing in cols if furnishing]
```

```python
cols
```

```python
def get_furnishing_count(details, furnishing):
    if isinstance(details, str):
        if f"No {furnishing}" in details:
            return 0
        pattern = re.compile(f"(\d+) {furnishing}")
        match = pattern.search(details)
        if match:
            return int(match.group(1))
        elif furnishing in details:
            return 1
    return 0
```

```python
for furnishing in cols:
    df3[furnishing] = df3['furnishDetails'].apply(lambda x: get_furnishing_count(x, furnishing))

# Create the new dataframe with the required columns
# furnishings_df = df[['furnishDetails'] + cols]
```

```python
df3.head()
```

```python
furnishings_df = df3[['furnishDetails'] + cols]
```

```python
furnishings_df.head()
```

```python
furnishings_df.drop(columns='furnishDetails',inplace=True)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(furnishings_df)
wcss_reduced = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_data)
    wcss_reduced.append(kmeans.inertia_)
```

```python
# Plot the results
plt.figure(figsize=(12, 8))
plt.plot(range(1,11), wcss_reduced, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal Number of Clusters (Reduced Range)')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
n_clusters = 3

# Fit the KMeans model
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(scaled_data)

# Predict the cluster assignments for each row
cluster_assignments = kmeans.predict(scaled_data)
```
