```python
df.drop(columns="link",inplace=True)
```

```python
df["BHK"]=df['property_name'].str.split('in').str.get(0).str[0]
```

```python
df['Place']=df['property_name'].str.split('in').str.get(1)
```

```python
df.dropna(subset=['price'],inplace=True)
```

```python
import re

def change(row):
    if 'Lac' in row:
        return float(re.findall(r'[\d.]*', row)[0])/100
    return re.findall(r'[\d.]*', row)[0]
```

```python
df['price']=df['price'].apply(change)
```

```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df['bedRoom']=df['bedRoom'].str.extract(r'([\d]+)')
df['bathroom']=df['bathroom'].str.extract(r'([\d]+)')
df['balcony']=df['balcony'].str.extract(r'([\d]+)')
df['area']=df['area'].str.extract(r'([\d]+)')
df['noOfFloor']=df['noOfFloor'].str.extract(r'([\d]+)')
```

```python


df['rate'] = df['rate'].str.extract(r'([\d,]+)')
df['rate'] = df['rate'].str.replace(',', '').astype(float)

```

```python
df.rename(columns={'rate':'price per squarefeet','noOfFloor':'total_Floors'},inplace=True)
```

```python
df.drop(columns=['property_name'],inplace=True)
```
