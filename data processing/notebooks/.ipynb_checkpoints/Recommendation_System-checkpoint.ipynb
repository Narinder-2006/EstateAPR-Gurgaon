```python
def recommend_properties(property_name, cosine_sim=cosine_sim1):
    # Get the index of the property that matches the name
    idx = df.index[df['PropertyName'] == property_name].tolist()[0]

    # Get the pairwise similarity scores with that property
    sim_scores = list(enumerate(cosine_sim1[idx]))

    # Sort the properties based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar properties
    sim_scores = sim_scores[1:6]

    # Get the property indices
    property_indices = [i[0] for i in sim_scores]
    
    recommendations_df = pd.DataFrame({
        'PropertyName': df['PropertyName'].iloc[property_indices],
        'SimilarityScore': sim_scores
    })

    # Return the top 10 most similar properties
    return recommendations_df
```

```python
import pandas as pd
import json

# Load the dataset
df_appartments = pd.read_csv('appartments.csv').drop(22)

# Function to parse and extract the required features from the PriceDetails column
def refined_parse_modified_v2(detail_str):
    try:
        details = json.loads(detail_str.replace("'", "\""))
    except:
        return {}

    extracted = {}
    for bhk, detail in details.items():
        # Extract building type
        extracted[f'building type_{bhk}'] = detail.get('building_type')

        # Parsing area details
        area = detail.get('area', '')
        area_parts = area.split('-')
        if len(area_parts) == 1:
            try:
                value = float(area_parts[0].replace(',', '').replace(' sq.ft.', '').strip())
                extracted[f'area low {bhk}'] = value
                extracted[f'area high {bhk}'] = value
            except:
                extracted[f'area low {bhk}'] = None
                extracted[f'area high {bhk}'] = None
        elif len(area_parts) == 2:
            try:
                extracted[f'area low {bhk}'] = float(area_parts[0].replace(',', '').replace(' sq.ft.', '').strip())
                extracted[f'area high {bhk}'] = float(area_parts[1].replace(',', '').replace(' sq.ft.', '').strip())
            except:
                extracted[f'area low {bhk}'] = None
                extracted[f'area high {bhk}'] = None

        # Parsing price details
        price_range = detail.get('price-range', '')
        price_parts = price_range.split('-')
        if len(price_parts) == 2:
            try:
                extracted[f'price low {bhk}'] = float(price_parts[0].replace('₹', '').replace(' Cr', '').replace(' L', '').strip())
                extracted[f'price high {bhk}'] = float(price_parts[1].replace('₹', '').replace(' Cr', '').replace(' L', '').strip())
                if 'L' in price_parts[0]:
                    extracted[f'price low {bhk}'] /= 100
                if 'L' in price_parts[1]:
                    extracted[f'price high {bhk}'] /= 100
            except:
                extracted[f'price low {bhk}'] = None
                extracted[f'price high {bhk}'] = None

    return extracted
# Apply the refined parsing and generate the new DataFrame structure
data_refined = []

for _, row in df_appartments.iterrows():
    features = refined_parse_modified_v2(row['PriceDetails'])
    
    # Construct a new row for the transformed dataframe
    new_row = {'PropertyName': row['PropertyName']}
    
    # Populate the new row with extracted features
    for config in ['1 BHK', '2 BHK', '3 BHK', '4 BHK', '5 BHK', '6 BHK', '1 RK', 'Land']:
        new_row[f'building type_{config}'] = features.get(f'building type_{config}')
        new_row[f'area low {config}'] = features.get(f'area low {config}')
        new_row[f'area high {config}'] = features.get(f'area high {config}')
        new_row[f'price low {config}'] = features.get(f'price low {config}')
        new_row[f'price high {config}'] = features.get(f'price high {config}')
    
    data_refined.append(new_row)

df_final_refined_v2 = pd.DataFrame(data_refined).set_index('PropertyName')
```

```python
df_final_refined_v2['building type_Land'] = df_final_refined_v2['building type_Land'].replace({'':'Land'})
```

```python
def recommend_properties_with_scores(property_name, top_n=246):
    
    # Get the similarity scores for the property using its name as the index
    sim_scores = list(enumerate(cosine_sim2[ohe_df_normalized.index.get_loc(property_name)]))
    
    # Sort properties based on the similarity scores
    sorted_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the indices and scores of the top_n most similar properties
    top_indices = [i[0] for i in sorted_scores[1:top_n+1]]
    top_scores = [i[1] for i in sorted_scores[1:top_n+1]]
    
    # Retrieve the names of the top properties using the indices
    top_properties = ohe_df_normalized.index[top_indices].tolist()
    
    # Create a dataframe with the results
    recommendations_df = pd.DataFrame({
        'PropertyName': top_properties,
        'SimilarityScore': top_scores
    })
    
    return recommendations_df

# Test the recommender function using a property name
recommend_properties_with_scores('M3M Golf Hills')
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import ast
import re
from difflib import SequenceMatcher

# Load your dataset


# Function to normalize distance to meters
def normalize_distance(dist_str):
    if pd.isna(dist_str) or dist_str == '':
        return np.nan
    
    dist_str = str(dist_str).lower().strip()
    match = re.search(r'(\d+\.?\d*)\s*(km|meter|m|kilometre)', dist_str, re.IGNORECASE)
    
    if match:
        value = float(match.group(1))
        unit = match.group(2).lower()
        
        if unit.startswith('k'):
            return value * 1000
        return value
    return np.nan

# Smart normalization that preserves meaning
def smart_normalize(name):
    """Normalize while preserving core identity"""
    name = str(name).lower().strip()
    
    # Remove only location/administrative suffixes that don't change identity
    # BUT keep the main name intact
    
    # Standardize common abbreviations
    name = re.sub(r'\brd\b', 'road', name)
    name = re.sub(r'\bst\b', 'street', name)
    name = re.sub(r'\bave\b', 'avenue', name)
    name = re.sub(r'\bsec\b', 'sector', name)
    name = re.sub(r'\bexpy\b', 'expressway', name)
    name = re.sub(r'\bhwy\b', 'highway', name)
    name = re.sub(r'\bintl\b', 'international', name)
    name = re.sub(r'\bint\b', 'international', name)
    
    # Remove punctuation but keep spaces
    name = re.sub(r'[,\.\'\"\(\)\/\:\;]', ' ', name)
    name = re.sub(r'\s+', ' ', name).strip()
    
    return name

# Check if two names are the same entity
def are_same_location(loc1, loc2):
    """
    Determine if two locations are the same using intelligent rules
    """
    # Basic preprocessing
    n1 = smart_normalize(loc1)
    n2 = smart_normalize(loc2)
    
    # Exact match after normalization
    if n1 == n2:
        return True
    
    # Extract core names (remove common suffixes/prefixes carefully)
    def extract_core(name):
        # Remove only trailing location qualifiers
        name = re.sub(r'\s+(gurgaon|gurugram|delhi|haryana|india)$', '', name)
        name = re.sub(r'\s+sector\s+\d+[a-z]?$', '', name)
        return name.strip()
    
    core1 = extract_core(n1)
    core2 = extract_core(n2)
    
    # If cores match exactly
    if core1 == core2 and core1:
        return True
    
    # Check if one is clearly an extension of the other
    # E.g., "Artemis Hospital" vs "Artemis Hospital Gurgaon"
    if len(core1) > 10 and len(core2) > 10:
        if core1 in core2 or core2 in core1:
            # Only if the difference is just location name
            diff_len = abs(len(n1) - len(n2))
            if diff_len < 20:  # Reasonable difference
                return True
    
    # Hospital/Bank/ATM specific rules
    # "HDFC Bank" should match "HDFC Bank ATM" but NOT "ICICI Bank"
    if 'bank' in n1 and 'bank' in n2:
        # Extract bank name (before "bank")
        bank1 = re.search(r'(\w+(?:\s+\w+)?)\s+bank', n1)
        bank2 = re.search(r'(\w+(?:\s+\w+)?)\s+bank', n2)
        if bank1 and bank2:
            if bank1.group(1) == bank2.group(1):
                # Same bank, one might have ATM
                if 'atm' in n1 or 'atm' in n2:
                    return True
    
    # Hospital specific - match only if same hospital name
    if 'hospital' in n1 and 'hospital' in n2:
        # Extract hospital name (before "hospital")
        hosp1 = re.sub(r'\s+hospital.*', '', n1).strip()
        hosp2 = re.sub(r'\s+hospital.*', '', n2).strip()
        if hosp1 and hosp2 and hosp1 == hosp2:
            return True
    
    # School specific - match only if same school name
    if 'school' in n1 and 'school' in n2:
        school1 = re.sub(r'\s+school.*', '', n1).strip()
        school2 = re.sub(r'\s+school.*', '', n2).strip()
        if school1 and school2 and school1 == school2:
            return True
    
    # Expressway/Highway variants
    if ('expressway' in n1 or 'highway' in n1) and ('expressway' in n2 or 'highway' in n2):
        # Extract the name before expressway/highway
        name1 = re.sub(r'\s+(expressway|highway).*', '', n1).strip()
        name2 = re.sub(r'\s+(expressway|highway).*', '', n2).strip()
        if name1 and name2 and name1 == name2:
            return True
    
    # NH (National Highway) special case
    if n1.startswith('nh') and n2.startswith('nh'):
        # Extract highway number
        num1 = re.search(r'nh\s*(\d+)', n1)
        num2 = re.search(r'nh\s*(\d+)', n2)
        if num1 and num2 and num1.group(1) == num2.group(1):
            return True
    
    # Metro station variants
    if 'metro' in n1 and 'metro' in n2:
        # Extract sector/area name
        base1 = re.sub(r'\s+(metro|rapid|station).*', '', n1).strip()
        base2 = re.sub(r'\s+(metro|rapid|station).*', '', n2).strip()
        if base1 and base2 and base1 == base2:
            return True
    
    # Very high similarity (95%+) for typos
    similarity = SequenceMatcher(None, n1, n2).ratio()
    if similarity >= 0.95:
        return True
    
    return False

# Filter useless locations
def should_filter_location(name):
    """Filter out very specific or useless locations"""
    name_lower = name.lower()
    
    filter_patterns = [
        r'unnamed',
        r'^\d+\s*bhk',
        r'plot\s*(no|number)?\s*\d+',
        r'flat\s*(no|number)?\s*\d+',
    ]
    
    for pattern in filter_patterns:
        if re.search(pattern, name_lower):
            return True
    
    if len(name.strip()) < 3:
        return True
    
    return False

# Parse LocationAdvantages
def parse_location_advantages(loc_str):
    if pd.isna(loc_str):
        return {}
    try:
        if isinstance(loc_str, str):
            return ast.literal_eval(loc_str)
        return loc_str
    except:
        return {}

print("="*70)
print("PROPERTY RECOMMENDATION SYSTEM - SMART DEDUPLICATION")
print("="*70)

print("\nStep 1: Parsing location data...")
df['LocationAdvantagesParsed'] = df['LocationAdvantages'].apply(parse_location_advantages)

# Collect all locations
print("\nStep 2: Collecting all locations...")
all_raw_locations = set()
for idx, row in df.iterrows():
    locations = row['LocationAdvantagesParsed']
    all_raw_locations.update(locations.keys())

print(f"Total raw location names: {len(all_raw_locations)}")

# Filter out useless locations
print("\nStep 3: Filtering out useless locations...")
filtered_locations = [loc for loc in all_raw_locations if not should_filter_location(loc)]
print(f"After filtering: {len(filtered_locations)} locations")
print(f"Filtered out: {len(all_raw_locations) - len(filtered_locations)} locations")

# Build canonical location map with smart deduplication
print("\nStep 4: Smart duplicate detection (this may take a moment)...")
location_to_canonical = {}
canonical_info = {}

sorted_locs = sorted(filtered_locations)

for i, loc in enumerate(sorted_locs):
    if i % 100 == 0:
        print(f"  Processing {i}/{len(sorted_locs)}...")
    
    # Check if similar to any existing canonical location
    found_match = False
    
    for canonical in list(canonical_info.keys()):
        if are_same_location(loc, canonical):
            location_to_canonical[loc] = canonical
            canonical_info[canonical]['variants'].append(loc)
            canonical_info[canonical]['count'] += 1
            found_match = True
            break
    
    if not found_match:
        # This is a new unique location
        location_to_canonical[loc] = loc
        canonical_info[loc] = {
            'canonical': loc,
            'variants': [loc],
            'count': 1
        }

# Get final unique locations
unique_locations = sorted(canonical_info.keys())
print(f"\nFinal unique locations: {len(unique_locations)}")
print(f"Duplicates merged: {len(filtered_locations) - len(unique_locations)}")

# Show merged duplicates
print("\n" + "="*70)
print("MERGED DUPLICATE LOCATIONS (showing those with 2+ variants):")
print("="*70)
sorted_by_count = sorted(canonical_info.items(), key=lambda x: x[1]['count'], reverse=True)
merge_count = 0
for canonical, info in sorted_by_count:
    if info['count'] > 1:
        merge_count += 1
        if merge_count <= 30:  # Show top 30
            print(f"\n{merge_count}. Canonical: '{canonical}'")
            print(f"   Variants ({info['count']} total):")
            for variant in info['variants']:
                if variant != canonical:
                    print(f"      → {variant}")

# Create distance matrix
print("\n" + "="*70)
print("Step 5: Creating property-location distance matrix...")
print("="*70)

distance_matrix = pd.DataFrame(index=df['PropertyName'], columns=unique_locations)

# Fill matrix with distances
for idx, row in df.iterrows():
    prop_name = row['PropertyName']
    locations = row['LocationAdvantagesParsed']
    
    for loc, dist in locations.items():
        if loc in location_to_canonical:
            canonical = location_to_canonical[loc]
            distance_meters = normalize_distance(dist)
            
            # Keep minimum distance if multiple entries
            current_dist = distance_matrix.loc[prop_name, canonical]
            if pd.isna(current_dist) or (not pd.isna(distance_meters) and distance_meters < current_dist):
                distance_matrix.loc[prop_name, canonical] = distance_meters

# Convert to numeric
distance_matrix = distance_matrix.apply(pd.to_numeric, errors='coerce')

print(f"\nDistance Matrix Shape: {distance_matrix.shape}")
print(f"Properties: {distance_matrix.shape[0]}")
print(f"Unique Locations: {distance_matrix.shape[1]}")

# Calculate sparsity
total_cells = distance_matrix.shape[0] * distance_matrix.shape[1]
non_null_cells = distance_matrix.notna().sum().sum()
sparsity = (1 - non_null_cells / total_cells) * 100

print(f"Non-null distances: {non_null_cells}")
print(f"Matrix sparsity: {sparsity:.2f}%")

# Save distance matrix
distance_matrix.to_csv('property_location_distance_matrix.csv')
print("\n✓ Distance matrix saved to 'property_location_distance_matrix.csv'")

# Save location mapping
mapping_data = []
for canonical, info in canonical_info.items():
    mapping_data.append({
        'Canonical_Location': canonical,
        'Variant_Count': info['count'],
        'All_Variants': ' | '.join(info['variants'])
    })

mapping_df = pd.DataFrame(mapping_data).sort_values('Variant_Count', ascending=False)
mapping_df.to_csv('location_deduplication_mapping.csv', index=False)
print("✓ Location mapping saved to 'location_deduplication_mapping.csv'")

# --- COSINE SIMILARITY ---
print("\n" + "="*70)
print("Step 6: Calculating cosine similarity...")
print("="*70)

# Prepare feature matrix (inverse distance as similarity)
feature_matrix = distance_matrix.fillna(0)
feature_matrix = 1 / (feature_matrix + 1)  # Inverse distance

# Calculate cosine similarity
similarity_matrix = cosine_similarity(feature_matrix)

# Create similarity DataFrame
similarity_df = pd.DataFrame(
    similarity_matrix,
    index=df['PropertyName'],
    columns=df['PropertyName']
)

print(f"Similarity Matrix Shape: {similarity_df.shape}")

# Save similarity matrix
similarity_df.to_csv('property_similarity_matrix.csv')
print("✓ Similarity matrix saved to 'property_similarity_matrix.csv'")

# --- RECOMMENDATIONS ---
def get_recommendations(property_name, similarity_df, top_n=5):
    """Get top N similar properties"""
    if property_name not in similarity_df.index:
        print(f"Property '{property_name}' not found!")
        return None
    
    similar_scores = similarity_df[property_name].sort_values(ascending=False)
    recommendations = similar_scores[similar_scores.index != property_name].head(top_n)
    
    return recommendations

# Example recommendations
if len(df) > 0:
    example_property = df['PropertyName'].iloc[0]
    print("\n" + "="*70)
    print(f"EXAMPLE: Top 5 Similar Properties to '{example_property}'")
    print("="*70)
    recommendations = get_recommendations(example_property, similarity_df, top_n=5)
    if recommendations is not None:
        for i, (prop, score) in enumerate(recommendations.items(), 1):
            print(f"{i}. {prop}")
            print(f"   Similarity: {score:.4f} ({score*100:.2f}%)")

# Final summary
print("\n" + "="*70)
print("FINAL SUMMARY")
print("="*70)
print(f"✓ Total Properties: {len(df)}")
print(f"✓ Raw Locations: {len(all_raw_locations)}")
print(f"✓ After Filtering: {len(filtered_locations)}")
print(f"✓ Final Unique Locations: {len(unique_locations)}")
print(f"✓ Duplicates Merged: {len(filtered_locations) - len(unique_locations)}")
print(f"✓ Distance Matrix: {distance_matrix.shape[0]} × {distance_matrix.shape[1]}")
print(f"✓ Matrix Sparsity: {sparsity:.2f}%")
print("\nGenerated Files:")
print("  1. property_location_distance_matrix.csv")
print("  2. property_similarity_matrix.csv")
print("  3. location_deduplication_mapping.csv")
print("="*70)
```
