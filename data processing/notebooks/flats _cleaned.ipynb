# Data Cleaning 

### Messy Data
##### "property name"- contain no of BHK and address
##### "society" - Rating and Society Name
### Dirty Data
##### 'price' - lacs and crore
##### 'area'- rename price per square feet.
##### 'bedroom'- no of bedrooms.
##### 'bathroom' - no of bathrooms.
##### 'balcony' - no of balcony.



```python
import pandas as pd
import re
```

```python
df=pd.read_csv(r"C:\Users\VICTUS\Downloads\flats - flats.csv")
```

```python
import pandas as pd





pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
df.drop(columns="link",inplace=True)
```

```python
df.head()
```

```python
df["BHK"]=df['property_name'].str.split('in').str.get(0).str[0]
```

```python
df.head()
```

```python
df['Place']=df['property_name'].str.split('in').str.get(1)
```

```python
df.head(5)
```

```python
df['price'].value_counts()
```

```python
df.dropna(subset=['price','area','floorNum'],inplace=True)
```

```python
df.info()
```

```python
import re

def change(row):
    if 'Lac' in row:
        return int(re.findall(r'\d+', row)[0])/100
    return re.findall(r'[\d.]*', row)[0]

    
```

```python
df['price']=df['price'].apply(change)
```

```python


df['area'] = df['area'].str.extract(r'([\d,]+)')
df['area'] = df['area'].str.replace(',', '').astype(float)

```

```python
df['bedRoom']=df['bedRoom'].str.extract(r'([\d]+)')
df['bathroom']=df['bathroom'].str.extract(r'([\d]+)')
df['balcony']=df['balcony'].str.extract(r'([\d]+)')
```

```python
df.rename(columns={'area':'price per squarefeet'},inplace=True)
```

```python
df
```

```python
df['total_Floors']=df['floorNum'].str.split().str.get(2)
```

```python
df['Floor of Flat']=df['floorNum'].str.split().str.get(0)
```

```python
df
```

```python
df['Floor of Flat']=df['Floor of Flat'].str.extract(r'([\d]+)')
```

```python
df.drop(columns=['property_name','floorNum'],inplace=True)
```

```python
df['society']
```

```python
import numpy as np
df['price'] = df['price'].replace('', np.nan)
df['price'] = df['price'].astype(np.float64)
```

```python
df.info()
```

```python
df['area']=df['price']*10000000/df['price per squarefeet']
```
